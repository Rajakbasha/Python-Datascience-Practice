```python
## list:- Lists are ordered collection of items which can be of different data types .Lists are mutablemeaning  their elements can be changed after they are defined .syntax: they are defined using square brackets [].
numbers=[1,2,3,4,5,6]
fruits=['apple','banana','orage','coconut']
#Accesing elements of a list 
first_number=numbers[0]
second_fruit="Fruits:",fruits[2]
print("Numbers:",numbers)
print("Fruit:",fruits)
print("First number:",first_number)
print("Second Fruit:",second_fruit)

```

```python
# 2. TUPLE :-Tuples are ordered collection of items which can be of different data types. tuples are  immutable 
# meaning their element cannot be changed once defined sysntax  they are defined  using parentheses ()
# DEFINING TUPLE 
point=(100,200,900)
dimensions=(100,200,300)
x=point[0]
y=point[1]
z=point[2]
print("Point:",point)
print("Dimensions:",dimensions)
print("X-coordinate:",x)
print("Y-coordinate:",y)
print("Z-coordinate:",z)
```

```python
# SET :- A set of is unordered collection of unique elements .it is iterable , mutable and does not allow duplicates elements 
# This makes sets unfit for indexing and slicing .syntax : Sets are created using  curly braces {} or by using the set () constructor.
# Creating a set 
my_set={1,2,3,4,5}
#Adding elements to a set 
my_set.add(6)
print(my_set)
```

```python
my_set.remove(3)
print(my_set)
```

```python
# SET OPERATORS 
set1={1,2,3}
set2={3,4,5}
print(set1.union(set2))
print(set1.intersection(set2))
```
