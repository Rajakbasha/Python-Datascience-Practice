# conditional operators 

 Python logical operators are used to combine conditional statements here are the three logical operators in python along with example 


```python
# AND (and): returs True if both condition on its left and right are True are False 
x=5
y=10
print(x<10 and y>5)
```

```python
x=500
y=100
print(x<100 and y<500)
```

```python
# OR (or) returns true if at least one of the conditons on its left or right is True 
x=5
y=10
print(x<10 or y<5)
```
